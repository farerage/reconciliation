```python
NIC = NIC.groupby(['Date']).agg({'MERCHANT_TRANS_ID':'count','TPV' : 'sum', 'MM Cost' :'sum', 'SETTLE_AMOUNT' : 'sum'})
```

```python
NIC.rename(columns = {'MERCHANT_TRANS_ID' : "#Trx"}, inplace = True)
```
