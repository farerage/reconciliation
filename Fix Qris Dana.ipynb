```python
data_compare = pd.merge(new_dana_qris_report,dana_qris_gds, left_on='MERCHANT_TRANS_ID_DASH', right_on='qr_id_GDS', how='outer', indicator=True)
```

```python
data_compare
```

```python
data_compare['Match?'] = np.where((data_compare['MERCHANT_TRANS_ID_DASH'] == data_compare['qr_id_GDS']) & (data_compare['TXN_AMOUNT_y_DASH'] == data_compare['transaction_amount_GDS']) , "Recon", "Unrecon")
```

```python
data_compare
```

```python
data_compare['_merge'] = data_compare['_merge'].cat.rename_categories({'left_only': 'Dash Only', 'right_only': 'GDS Only'})
```

```python
#data_compare.rename(columns={"last_updated_datetime": "Date"}, inplace=True)
```

```python
unrecon = data_compare[data_compare['Match?'] == 'Unrecon']
```

```python
indexunrecon = data_compare[data_compare['Match?'] == 'Unrecon'].index
```

```python
data_compare = data_compare.drop(indexunrecon)
```

```python
with pd.ExcelWriter("Dana Qris 31 Jan 24 New Template.xlsx") as writer:
    dana_qris_gds_raw.to_excel(writer, sheet_name="GDS Raw", index=False)
    dana_qris_report_raw.to_excel(writer, sheet_name="Dash Report Raw", index=False)
    data_compare.to_excel(writer, sheet_name="Line Per Line Recon", index=False)
    unrecon.to_excel(writer, sheet_name="Unrecon", index=False)
    dash_duplicated.to_excel(writer, sheet_name="Multiple Dash", index=False)
    gds_duplicated.to_excel(writer, sheet_name="Multiple GDS", index=False)
```
