```python
dana_report = dana_report.drop(['Time', 'SOURCE', 'MERCHANT_NAME', 'Cheque Number', 'PAYER_ROLE_ID','Reference7', 'Reference8', 'Reference9','FUND_AMOUNT_CURRENCY'], axis = 1)
```

```python
dana_report = dana_report.add_suffix('_MUTASI')
```

```python
gds_dana_report = gds_dana_report.drop(['Source Account', 'Domain','testing_account','Tx Ref Number','Vendor Ref ID', 'Bank Code', 'Bank Account' ], axis = 1)
```

```python
gds_dana_report = gds_dana_report.add_suffix('_GDS')
```

Start rekon line per line

```python
data_compare = pd.merge(dana_report,gds_dana_report, left_on='REQUEST_ID.1_MUTASI', right_on='Reference Number_GDS', how='outer', indicator=True)
```

```python
data_compare
```

```python
data_compare['Match?'] = np.where((data_compare['REQUEST_ID.1_MUTASI'] == data_compare['Reference Number_GDS']) & (data_compare[' FUND_AMOUNT _MUTASI'] == data_compare['Amount_GDS']) , "Recon", 'Unrecon')
```

```python
data_compare
```

```python
data_compare['_merge'] = data_compare['_merge'].cat.rename_categories({'left_only': 'Dash Only', 'right_only': 'GDS Only'})
```

rekon line per line done

```python
data_compare
```

```python
unrecon = data_compare[data_compare['Match?'] == 'Unrecon']
```

```python
index_unrecon = data_compare[data_compare['Match?'] == 'Unrecon'].index
```

```python
data_compare = data_compare.drop(index_unrecon)
```

```python
data_compare
```

```python
with pd.ExcelWriter("Dana Disburse 1-7 Mar 24 New Template.xlsx") as writer:
    gds_dana_report_raw.to_excel(writer, sheet_name="GDS Raw", index=False)
    dana_report_raw.to_excel(writer, sheet_name="Mutasi Report Raw", index=False)
    data_compare.to_excel(writer, sheet_name="Line Per Line Recon", index=False)
    unrecon.to_excel(writer, sheet_name="Unrecon", index=False)
```

```python
print("Done")
```
