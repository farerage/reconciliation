```python
data_dash.rename(columns={"Merchant Transaction Number": "MTN"}, inplace=True)
```

```python
data_gds = data_gds.drop(['settlement_time', 'charge_vendor_code','merchant_id', 'Transaction SN'],axis = 1)
```

```python
data_dash = data_dash.drop(['Diterima Dari', 'MPAN', 'MID','NMID','TID','Tipe','Kode Persetujuan','No. Referensi Pengambilan','Kode Transaksi','Catatan','Extended Merchant Trx No','Loyalty Number'], axis=1)
```

```python
data_dash[data_dash.duplicated()]
```

```python
dash_duplicated = data_dash[data_dash.duplicated()]
```

```python
data_dash = data_dash.drop_duplicates()
```

```python
data_gds[data_gds.duplicated()]
```

```python
gds_duplicated = data_gds[data_gds.duplicated()]
```

```python
data_gds = data_gds.drop_duplicates()
```

```python
data_dash = data_dash.add_suffix('_DASH')
```

```python
data_gds = data_gds.add_suffix('_GDS')
```

```python
data_compare = pd.merge(data_dash,data_gds, left_on='MTN_DASH', right_on='qr_id_GDS', how='outer', indicator=True)
```

```python
data_compare['Match?'] = np.where((data_compare['MTN_DASH'] == data_compare['qr_id_GDS']) & (data_compare['Nominal Transaksi (Rp)_DASH'] == data_compare['transaction_amount_GDS']) , "Recon", "Unrecon")
```

```python
data_compare
```

```python
data_compare['_merge'] = data_compare['_merge'].cat.rename_categories({'left_only': 'Dash Only', 'right_only': 'GDS Only'})
```

```python
data_compare
```

```python
unrecon = data_compare[data_compare['Match?'] == 'Unrecon']
```

```python
index_unrecon = data_compare[data_compare['Match?'] == 'Unrecon'].index
```

```python
data_compare = data_compare.drop(index_unrecon)
```

```python
unrecon.dtypes
```

```python
with pd.ExcelWriter("Qris Nobu 10 Mar 24 New Template.xlsx") as writer:
    data_gds_raw.to_excel(writer, sheet_name="GDS Raw", index=False)
    data_dash_raw.to_excel(writer, sheet_name="Dash Report Raw", index=False)
    data_compare.to_excel(writer, sheet_name="Line Per Line Recon", index=False)
    unrecon.to_excel(writer, sheet_name="Unrecon", index=False)
    dash_duplicated.to_excel(writer, sheet_name="Multiple MTN Dashboard", index=False)
    gds_duplicated.to_excel(writer, sheet_name="Multiple MTN GDS", index=False)
```

```python
NIC = data_dash_raw[['Tanggal', 'Penerbit', 'Merchant Transaction Number','Nominal Transaksi (Rp)','Biaya MDR (Rp)','Jumlah (Rp)']]
```

```python
NIC.rename(columns={"Tanggal": "Date", 'Nominal Transaksi (Rp)' : 'TPV', 'Biaya MDR (Rp)' : 'MM Cost', 'Jumlah (Rp)' : 'Settlement Amount', 'Penerbit' : 'Channel Type'}, inplace=True)
```

```python
pd.options.mode.copy_on_write = True
```

```python
NIC[NIC['Channel Type'] == 'Nobu']
```

```python
offus = NIC[NIC['Channel Type'] != 'Nobu'].index
```

```python
NIC['Rev Sharing'] = np.nan
```

```python
NIC.loc[offus, 'Rev Sharing'] = NIC['MM Cost'] - NIC['TPV'] * 0.00189
```

```python
onus = NIC[NIC['Rev Sharing'].isna()].index
```

```python
NIC.loc[onus, 'Rev Sharing'] = 0
```

```python
NIC['Rev Sharing'] = NIC['Rev Sharing'].round(2)
```

```python
NIC.loc[offus, 'Channel Type'] = "Off Us"
```

```python
onuscek = data_dash_raw[data_dash_raw['Penerbit'] == 'Nobu'].index
```

```python
NIC.loc[onuscek, 'Channel Type'] = "On Us"
```

```python
NIC
```

```python
NIC = NIC.groupby(['Date', 'Channel Type']).agg({'Merchant Transaction Number':'count','TPV' : 'sum', 'MM Cost' :'sum', 'Settlement Amount' : 'sum', 'Rev Sharing' : 'sum' })
```

```python
NIC.rename(columns = {'Merchant Transaction Number' : "#Trx"}, inplace = True)
```

```python
with pd.ExcelWriter("NIC Qris Nobu.xlsx") as writer:
    NIC.to_excel(writer, sheet_name="NIC Qris Nobu", index=True) #notes : using this code only for the first date 
```
